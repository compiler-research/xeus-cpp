# Interpreting the C programming language
# You can define functions, structs, etc ...

## Functions

```c++
double sqr(double a)
{
    return a * a;
}
```

```c++
double a = 2.5;
double asqr;
asqr = sqr(a);
printf("%lf\n", asqr);
```

## Structs

```c++
struct foo
{
    double value;
};

void foo_print(struct foo f)
{
    printf("Foo value = %lf\n",  f.value);
}
```

```c++
struct foo bar = {.value = 1.2};
foo_print(bar);
```

# Taking input from the user

```c++
#include <string.h>

char name[BUFSIZ];
if (fgets(name, sizeof(name), stdin) != NULL)
{
    name[strcspn(name, "\n")] = '\0';
}
```

```c++
printf("Your name is %s\n", name);
```
